```python
import pandas as pd
import yfinance as yf
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
stock_names = 1
index = '^NYA'
sic = 'NIO, FB, APPL'
```

```python
periodicidad = '2y'
intervalo = '1d'
factor = 252

df_index = yf.download( index , period = periodicidad , interval = intervalo)
df_index = df_index['Close']

df_sic = yf.download( sic , period = periodicidad , interval = intervalo)
df_sic = df_sic['Close']
```

```python
def rend(df):
    return np.log(df / df.shift(1)).dropna()

def cov(df):
    return df.cov()

rend_sic = rend(df_sic)
rend_index = rend(df_index)
df_help = rend_sic.join(rend_index) 
cov_index = cov(df_help)
bet = cov_index.iloc[:-1,-1] / np.var(rend_index)                
bet = pd.DataFrame(bet) 

ds = {'Rendimiento' : rend_sic.mean()*np.sqrt(factor)*100}
resul = pd.DataFrame(ds)                                    
resul['Var'] = np.var(rend_sic)                           
resul['Risk'] = np.sqrt(resul['Var']) *100                       
resul['Beta'] = bet
```
