```python
import pandas as pd
import yfinance as yf
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
ipc = 'SXRJ.DE AC.MX ALFA.MX ALSEA.MX AMXL.MX ASURB.MX BBAJIOO.MX BIMBOA.MX BOLSAA.MX CEMEXCPO.MX CUERVO.MX ELEKTRA.MX FEMSAUBD.MX GAPB.MX GCARSOA1.MX GCC.MX GFINBURO.MX GFNORTEO.MX GMEXICOB.MX GRUMAB.MX KIMBERA.MX KOFUBL.MX LABB.MX LIVEPOLC-1.MX MEGACPO.MX OMAB.MX ORBIA.MX PE&OLES.MX PINFRA.MX Q.MX RA.MX SITESB-1.MX TLEVISACPO.MX VESTA.MX VOLARA.MX WALMEX.MX'
```

```python
periodicidad = '1Y'
intervalo = '1d'
factor = 252

bmv = yf.download( ipc , period = periodicidad , interval = intervalo)
bmv = bmv['Close']

bmv_vol = yf.download( ipc , period = periodicidad , interval = intervalo)
bmv_vol = bmv_vol['Close']
```

```python
bmv.head(2)
```

```python
# Top 10 semanal
rend_semanal = pd.DataFrame()
rend_semanal['Rendimiento semanal'] = (np.log(bmv.iloc[-1] / bmv.iloc[-6])*100).round(2)
rend_semanal['Precio hace 5 d√≠as'] = bmv.iloc[-6].round(2)
rend_semanal['Ultimo precio'] = bmv.iloc[-1].round(2)
rend_semanal['Volumen promedio semanal'] = bmv_vol.iloc[-6:-1].mean().round(0)
rend_semanal.sort_values('Rendimiento semanal',ascending=False).head(10)
```

```python
px.line(bmv)
```

```python
top_10 = rend_semanal.sort_values('Rendimiento semanal',ascending=False).head(10)
graf = bmv.loc[:,top_10.index]
px.line(graf)
```

```python
print('El top 10 son: ', list(top_10.index))
```
