```python
import pandas as pd
import yfinance as yf
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
portafolio_names = 5
index = '^MXX'
port_nac =  'ALSEA.MX CUERVO.MX LABB.MX PE&OLES.MX PINFRA.MX'
port_int = 'NIO TSLA'
```

```python
periodicidad = '1y'
intervalo = '1d'
factor = 252

df_index = yf.download( index , period = periodicidad , interval = intervalo)
df_index = df_index['Close']

df_nac = yf.download( port_nac , period = periodicidad , interval = intervalo)
df_nac = df_nac['Close']

df_int = yf.download( port_int , period = periodicidad , interval = intervalo)
df_int = df_int['Close']

dolar = yf.download( 'MXN=X' , period = periodicidad , interval = intervalo)
dolar = dolar['Close']
```

```python
print('Nacionales: ', df_nac.keys())
print('Internacional: ', df_int.keys())
```

```python
# NACIONAL
x = input('Introduce un ticker ') + '.MX'
data = df_nac[x]
analysis = pd.DataFrame(data)
analysis['MA05'] = analysis[x].rolling(5).mean()
analysis['MA20'] = analysis[x].rolling(20).mean()
analysis['MAX'] = df_nac[x].max()
analysis['MIN'] = df_nac[x].min()
analysis['MEAN'] = df_nac[x].mean()
px.line(analysis)
```

```python
print('Acción: ', x)
print('Precio máximo: ',max(df_nac[x]))
print('Precio mínimo: ',min(df_nac[x]))
print('Precio ayer: ',df_nac[x].iloc[-1])
```

```python
# INTERNACIONAL
y = input('Introduce un ticker ')
data = df_int[y]
analysis = pd.DataFrame(data)
analysis['MA05'] = analysis[y].rolling(5).mean()
analysis['MA20'] = analysis[y].rolling(20).mean()
analysis['MAX'] = df_int[y].max()
analysis['MIN'] = df_int[y].min()
analysis['MEAN'] = df_int[y].mean()
px.line(analysis)
```

```python
print('Acción: ', y)
print('Precio máximo: ',max(df_int[y]))
print('Precio mínimo: ',min(df_int[y]))
print('Precio ayer: ',df_int[y].iloc[-1])
```

# Precios objetivo

```python
hl = df_nac.join(df_int).iloc[:,:-1]
```

```python
p = {'Acciones' : ['NIO', 'ALSEA', 'CUERVO', 'LABB', 'PE&OLES', 'PINFRA'] , 
     'Precio adquisición' : [854.78, 44.25, 48, 19.70, 293.41, 151.55],
     'Numero de acciones' : [3, 15, 20, 20, 1, 4]}
o = pd.DataFrame(p)
o['Precio objetivo mínimo'] = (o['Precio adquisición']*1.0025).round(2)
o.set_index('Acciones')
```

```python
z = hl.iloc[-1]
pos = pd.DataFrame({'Ultimo precio' : z}).round(2)
pos['Precio adquirido'] = [44.25, 48, 19.70, 293.41, 151.55, 854.78/dolar[-1]]
pos['# Acciones'] = [15, 20, 20, 1, 4, 3]
pos['Dif $'] = (pos['Ultimo precio'] - pos['Precio adquirido']).round(2)
pos['Dif %'] = ((pos['Dif $'] / pos['Precio adquirido'])*100).round(2)
pos['Ganancia/Perdida total'] = pos['Dif $'] * pos['# Acciones']
pos
```

```python
print('La ganancia o pérdida del portafolio es:', pos['Ganancia/Perdida total'].sum().round(2))
```
